```python
numeric_data = data.select_dtypes(include=['number'])

Q1 = numeric_data.quantile(0.25)
Q3 = numeric_data.quantile(0.75)
IQR = Q3 - Q1

# Lọc outlier cho phần dữ liệu số
filtered_numeric = numeric_data[~((numeric_data < (Q1 - 1.5 * IQR)) | (numeric_data > (Q3 + 1.5 * IQR)))]
# Ghép lại với cột chữ
data = filtered_numeric.join(data.select_dtypes(exclude=['number']))
```
